```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("Ecommerce_Consumer_Behavior_Analysis_Data.csv")
```

```python
df.head(10)
```

```python
df.isna().sum()
```

```python
df["Engagement_with_Ads"].value_counts()
```

```python
engagement = df.groupby("Engagement_with_Ads")["Customer_ID"].count()
valor = engagement.mean()
df["Engagement_with_Ads"].fillna("Medium", inplace=True)
```

```python
df.isna().sum()
```

```python
df["Social_Media_Influence"].value_counts()
```

```python
social = df.groupby("Social_Media_Influence")["Customer_ID"].count()
valor = social.mean()
df["Social_Media_Influence"].fillna("Low", inplace=True)
```

```python
df.isna().sum()
```

```python
genero_compra = df.groupby("Gender")["Frequency_of_Purchase"].count().sort_values(ascending=False)
sns.barplot(x = genero_compra.index, y = genero_compra.values, hue = genero_compra.index)
plt.xticks(rotation = 45)
plt.show()
```
